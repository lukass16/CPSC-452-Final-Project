# **View Meshes**

```python
import os
import trimesh
```

**Setup paths**

```python
# define category IDs
CANS = "02946921"
PLANES = "02691156"
CARS = "02958343"
```

```python
CATEGORY_ID = CARS

relative_data_path = "" # folder relative to script directory where the category folder is located
category_path = os.path.join(os.getcwd(), relative_data_path, CATEGORY_ID)

files = os.listdir(category_path)
```

```python
ID = 5 # set ID for mesh to view

# get mesh path
model_path = files[ID]
mesh_path = os.path.normpath(os.path.join(category_path, model_path, "models", "model_normalized.obj"))

# load the mesh
mesh = trimesh.load(mesh_path, force='mesh')

# display mesh
print("Mesh Summary:")
print(mesh)
mesh.show()
```
